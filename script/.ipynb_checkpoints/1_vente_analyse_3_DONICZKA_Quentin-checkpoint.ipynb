```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from utils.etoile_ventes_avis import graphe_etoile_ventes_avis

DATA_PATH = "../donnees/ecommerce/"
```

```python
# 0 - Chargement et préparation des données

sns.set()

F_VENTES_AVIS = pd.read_parquet("f_ventes_avis.parquet")
df = F_VENTES_AVIS.copy()

if "purchase_timestamp" in df.columns and not np.issubdtype(df["purchase_timestamp"].dtype, np.datetime64):
    df["purchase_timestamp"] = pd.to_datetime(df["purchase_timestamp"])

if "purchase_timestamp" in df.columns:
    df["annee_mois"] = df["purchase_timestamp"].dt.to_period("M").astype(str)

if "livree" in df.columns and "estimee" in df.columns:
    df["retard_h"] = df["livree"] - df["estimee"]
else:
    df["retard_h"] = np.nan

def classify_retard(h):
    if pd.isna(h):
        return np.nan
    if h <= -24:
        return "En avance (> 24h)"
    elif -24 < h <= 24:
        return "À l'heure (±24h)"
    elif 24 < h <= 72:
        return "En retard (1–3j)"
    else:
        return "Très en retard (> 3j)"

df["classe_retard"] = df["retard_h"].apply(classify_retard)

print("Shape df :", df.shape)
print("Colonnes principales :", [c for c in df.columns if c in ["value","score","livree","estimee","retard_h","classe_retard","annee_mois"]])
```

```python
# Analyse 3 : Comportement de paiement & impact business / satisfaction
# Cette analyse s’intéresse aux modes de paiement et au nombre de versements afin
# d’évaluer leur influence sur la valeur générée et sur la satisfaction client.
# Après avoir étudié les effets du temps et de la logistique, l’objectif est d’explorer
# d’autres facteurs potentiels qui pourraient expliquer certaines variations des avis.
```

```python
# Hypothèse 3
# Étant donné que l’analyse 2 a montré que les délais logistiques expliquent une grande
# partie de l’insatisfaction, on cherche maintenant à voir s’il existe d’autres causes
# indépendantes des retards. On suppose que certains comportements de paiement
# (paiement en plusieurs fois, méthodes spécifiques, montants plus élevés)
# pourraient être associés à un risque plus fort d’insatisfaction. L’idée est de vérifier
# si le type de paiement ou la structure du panier joue un rôle complémentaire dans la
# perception du service.
```

```python
# 1 - Répartition du CA total par mode de paiement

# Préparation
value_cols = [c for c in df.columns if c.startswith("value_")]

df1 = df[value_cols].sum().reset_index()
df1.columns = ["col", "total_value"]
df1["mode_paiement"] = df1["col"].str.replace("value_", "", regex=False)

# Affichage
plt.figure(figsize=(10, 5))
sns.barplot(data=df1, x="mode_paiement", y="total_value")
plt.xticks(rotation=45)
plt.title("Répartition du CA total par mode de paiement")
plt.xlabel("Mode de paiement")
plt.ylabel("CA total")
plt.tight_layout()
plt.show()
```

```python
# Le graphique montre que la carte de crédit domine largement le chiffre d’affaires total.
# Cela indique que la majorité des clients préfèrent ce mode de paiement pour leurs achats.
# Le boleto arrive en deuxième position mais reste très loin derrière, avec un usage beaucoup plus limité.
# Les autres moyens de paiement (débit, voucher, not_defined) génèrent très peu de CA.
# Cette répartition suggère une forte dépendance de l’entreprise au paiement par carte de crédit.
# Cela peut refléter une recherche de flexibilité des clients via les paiements en plusieurs fois.
# On observe donc un comportement de paiement très concentré sur une seule méthode principale.
```

```python
# 2 - Nombre moyen de versements selon le mode de paiement principal

# Préparation
value_cols = [c for c in df.columns if c.startswith("value_")]
int_cols = [c for c in df.columns if c.startswith("int_")]

df["main_payment_method"] = (
    df[value_cols]
    .apply(lambda row: row.idxmax() if row.notna().any() else np.nan, axis=1)
    .str.replace("value_", "", regex=False)
)

# Création d'une variable "total_versements"
if "total_versements" not in df.columns and int_cols:
    df["total_versements"] = df[int_cols].sum(axis=1)

df2 = (
    df.dropna(subset=["main_payment_method"])
      .groupby("main_payment_method")["total_versements"]
      .mean()
      .reset_index()
)

# Affichage
plt.figure(figsize=(10, 5))
sns.barplot(data=df2, x="main_payment_method", y="total_versements")
plt.xticks(rotation=45)
plt.title("Nombre moyen de versements par mode de paiement principal")
plt.xlabel("Mode de paiement principal")
plt.ylabel("Versements moyens")
plt.tight_layout()
plt.show()
```

```python
# Le graphique montre que la carte de crédit est le mode de paiement où les clients utilisent le plus de versements.
# Avec plus de 3,5 versements en moyenne, c’est le seul moyen qui permet vraiment d’étaler le paiement dans le temps.
# Les autres modes comme le boleto, la carte de débit ou les paiements “not_defined” restent proches d’un paiement en une seule fois.
# Cela montre qu’ils sont surtout utilisés pour des paiements comptants, sans possibilité de fractionner.
# Le voucher se situe un peu au-dessus, mais reste très loin derrière la carte de crédit.
# Cela confirme que le paiement en plusieurs fois repose presque uniquement sur la carte de crédit.
# On peut aussi imaginer que les clients utilisant ce mode ont des paniers plus élevés ou cherchent plus de flexibilité.
# Le graphique met donc bien en évidence deux comportements distincts : payer comptant ou payer en plusieurs fois.
```

```python
# 3 - Score moyen des avis par mode de paiement principal (avec nombre de commandes affiché)

# Préparation
value_cols = [c for c in df.columns if c.startswith("value_")]
if "main_payment_method" not in df.columns and value_cols:
    df["main_payment_method"] = df[value_cols].idxmax(axis=1).str.replace("value_", "", regex=False)

df3 = (
    df.dropna(subset=["main_payment_method", "score"])
      .groupby("main_payment_method")
      .agg(
          score_moyen=("score", "mean"),
          n_cmd=("score", "count")
      )
      .reset_index()
)

# Affichage
plt.figure(figsize=(10, 5))
ax = sns.barplot(data=df3, x="main_payment_method", y="score_moyen")

for i, row in df3.iterrows():
    ax.text(
        i,
        row["score_moyen"] + 0.05,
        f"n={row['n_cmd']}",
        ha="center",
        fontsize=9
    )

plt.xticks(rotation=45)
plt.title("Score moyen des avis selon le mode de paiement principal\n(avec nombre de commandes)")
plt.xlabel("Mode de paiement principal")
plt.ylabel("Score moyen")
plt.ylim(1, 5.2)
plt.tight_layout()
plt.show()
```

```python
# Les scores moyens sont quasiment identiques pour les modes de paiement vraiment utilisés.
# Cela montre que le mode de paiement n’a pas d’impact notable sur la satisfaction client.
# Le voucher, le boleto, la carte de crédit et la carte de débit obtiennent tous des scores autour de 4.
# La seule exception est “not_defined”, mais avec seulement 3 commandes, ce résultat n’est pas fiable.
# Globalement, on peut donc conclure que la satisfaction reste stable quel que soit le moyen de paiement choisi.
```

```python
# 4 - Lien entre montant payé et score d’avis
# Préparation
df4 = df[["value", "score"]].dropna().copy()

df4["tranche_montant"] = pd.cut(
    df4["value"],
    bins=[0, 100, 300, 600, 1500, df4["value"].max()],
    labels=[
        "0–100€",
        "100–300€",
        "300–600€",
        "600–1500€",
        "1500€+"
    ],
    include_lowest=True
)

pivot = (
    pd.crosstab(df4["tranche_montant"], df4["score"], normalize="index")
    .round(3)
)

# Affichage
plt.figure(figsize=(10, 7))
sns.heatmap(
    pivot,
    annot=True,
    cmap="Blues",
    fmt=".2f",
    cbar_kws={"label": "Proportion d'avis"}
)
plt.title("Distribution des scores par tranche de montant payé")
plt.xlabel("Score d'avis")
plt.ylabel("Tranche de montant payé")
plt.tight_layout()
plt.show()
```

```python
# Le graphique met surtout en évidence que les avis 1 étoile augmentent fortement quand le montant de la commande devient élevé.
# Pour les petits achats, les avis négatifs restent assez rares, mais ils montent a plus du double pour les commandes supérieures à 1500€.
# Cela montre clairement que les clients sont beaucoup moins tolérants quand ils dépensent une grosse somme.
# Un retard, un produit abîmé ou un service décevant pèse plus lourd quand la facture est importante.
# Malgré cela, la majorité des avis restent positifs, mais les commandes chères génèrent plus de réactions extrêmes.
# On a donc un schéma logique : plus le montant est élevé, plus le risque d’insatisfaction augmente.
# Ce comportement reflète des attentes plus fortes et une sensibilité accrue au moindre problème.
# Le montant payé semble donc jouer un rôle réel dans la probabilité de recevoir un avis très négatif.
```

```python
# 5 - Taux d’avis négatifs (score 1–2) par mode de paiement principal

# Préparation
value_cols = [c for c in df.columns if c.startswith("value_")]
if "main_payment_method" not in df.columns and value_cols:
    df["main_payment_method"] = (
        df[value_cols]
        .idxmax(axis=1)
        .str.replace("value_", "", regex=False)
    )

df_temp = df.dropna(subset=["main_payment_method", "score"]).copy()

# Retrait explicite de "not_defined"
df_temp = df_temp[df_temp["main_payment_method"] != "not_defined"]

df_temp["negatif"] = df_temp["score"].isin([1, 2])

df5 = (
    df_temp.groupby("main_payment_method")["negatif"]
           .mean()
           .reset_index()
)
df5["taux_negatif"] = df5["negatif"] * 100

# Affichage
plt.figure(figsize=(10, 5))
sns.barplot(data=df5, x="main_payment_method", y="taux_negatif")
plt.xticks(rotation=45)
plt.title("Taux d’avis négatifs (1–2) par mode de paiement principal")
plt.xlabel("Mode de paiement principal")
plt.ylabel("Taux d’avis négatifs (%)")
plt.ylim(0, df5["taux_negatif"].max() * 1.15)  # marge pour lisibilité
plt.tight_layout()
plt.show()
```

```python
# Le graphique montre que le voucher a un taux d’avis négatifs un peu plus élevé que les autres modes de paiement, d’environ 3 %.
# Comme le voucher correspond souvent à un bon de réduction ou un geste commercial, il peut être utilisé par des clients ayant déjà rencontré un problème auparavant.
# Cela peut expliquer en partie pourquoi leurs avis sont légèrement plus négatifs que la moyenne.
# L’écart reste modéré, mais il peut aussi venir d’un volume de commandes plus faible, ce qui rend les variations plus visibles.
# Les autres modes de paiement ont des résultats très proches, ce qui montre que le moyen de paiement n’influence pas vraiment la satisfaction globale.
# Cette petite différence sur le voucher doit donc être interprétée avec prudence et replacée dans son contexte.
# Elle peut refléter le profil des clients qui l’utilisent plutôt qu’un réel problème lié au paiement.
# Globalement, le graphique suggère que la satisfaction dépend surtout d’autres facteurs, comme la livraison ou la qualité du produit.
```

```python
# 6 - Montant moyen du panier selon le nombre de versements

# Préparation
int_cols = [c for c in df.columns if c.startswith("int_")]

if "nb_versements" not in df.columns and int_cols:
    df["nb_versements"] = df[int_cols].sum(axis=1)

df6 = (
    df.dropna(subset=["nb_versements", "value"])
      .groupby("nb_versements")
      .agg(
          montant_moyen=("value", "mean"),
          nb_commandes=("value", "count")
      )
      .reset_index()
)

# Affichage
plt.figure(figsize=(10, 5))
sns.lineplot(data=df6, x="nb_versements", y="montant_moyen", marker="o")
plt.title("Montant moyen du panier selon le nombre de versements")
plt.xlabel("Nombre de versements")
plt.ylabel("Montant moyen du panier (€)")
plt.tight_layout()
plt.show()
```

```python
# Le graphique montre que plus le nombre de versements augmente, plus le montant moyen du panier tend à être élevé.
# Les clients qui paient comptant ou avec peu de versements ont des paniers plutôt faibles, souvent en dessous de 200 €.
# Dès que l’on dépasse une dizaine de versements, le montant moyen grimpe nettement, passant régulièrement au-dessus de 400 €.
# On voit aussi plusieurs pics importants, notamment autour de 20 à 25 versements où les paniers peuvent dépasser les 600 €.
# Cela indique que les paiements en plusieurs fois sont surtout utilisés pour des achats plus coûteux.
# Les fluctuations présentes dans le graphique peuvent être liées à des volumes plus faibles pour les versements très élevés.
# Globalement, le lien est clair : les clients choisissent d’étaler le paiement quand le montant devient important.
# Ce comportement confirme que les versements servent surtout à rendre accessibles des achats plus chers.
```

```python
# 7 - Retard moyen de livraison selon le nombre de versements

# Préparation
int_cols = [c for c in df.columns if c.startswith("int_")]
if "installments_total" not in df.columns and int_cols:
    df["installments_total"] = df[int_cols].sum(axis=1)

df7 = (
    df.dropna(subset=["installments_total", "retard_h"])
      .groupby("installments_total")["retard_h"]
      .mean()
      .reset_index()
      .rename(columns={"installments_total": "versements"})
)

# Affichage
plt.figure(figsize=(10, 5))
sns.lineplot(data=df7, x="versements", y="retard_h", marker="o")
plt.title("Retard moyen de livraison selon le nombre de versements")
plt.xlabel("Nombre de versements")
plt.ylabel("Retard moyen (heures)")
plt.axhline(0, linestyle="--")
plt.tight_layout()
plt.show()

```

```python
# Le graphique montre que les commandes sont en moyenne livrées en avance, quel que soit le nombre de versements.
# Après un point isolé légèrement en retard pour les paiements comptants, les retards deviennent négatifs et assez stables autour de -250 à -300 heures.
# Cela signifie que les livraisons arrivent généralement plus tôt que la date estimée, même pour des achats fractionnés.
# On observe toutefois quelques variations importantes quand le nombre de versements dépasse 15, probablement dues à un faible nombre de commandes.
# Ces pics extrêmes doivent donc être interprétés avec prudence, car ils reflètent sans doute des cas particuliers.
# Globalement, il n’y a pas de lien clair entre le nombre de versements et les retards de livraison.
# Les clients qui paient en plusieurs fois ne semblent donc pas subir de délais plus longs.
# Le graphique suggère que la logistique reste globalement performante quelle que soit la manière de payer.
```

```python
# 8 - Heatmap retard moyen + volume de commandes selon période et nombre de versements

# Préparation
int_cols = [c for c in df.columns if c.startswith("int_")]

if "installments_total" not in df.columns and int_cols:
    df["installments_total"] = df[int_cols].sum(axis=1)

df_heat = df.dropna(subset=["installments_total", "retard_h", "annee_mois"]).copy()
df_heat = df_heat[df_heat["installments_total"] <= 25]

agg = (
    df_heat
    .groupby(["annee_mois", "installments_total"])
    .agg(
        retard_moyen=("retard_h", "mean"),
        commandes=("order_id", "count")
    )
    .reset_index()
)

pivot_retard = agg.pivot(index="annee_mois", columns="installments_total", values="retard_moyen")
pivot_count  = agg.pivot(index="annee_mois", columns="installments_total", values="commandes")

plt.figure(figsize=(16, 10))
sns.heatmap(pivot_retard, cmap="coolwarm", center=0, annot=pivot_count, fmt=".0f")
plt.title("Retard moyen (couleur) et nombre de commandes (texte) selon période et versements")
plt.xlabel("Nombre de versements")
plt.ylabel("Période (année-mois)")
plt.tight_layout()
plt.show()
```

```python
# En regardant la heatmap, on retrouve les deux périodes déjà repérées auparavant, notamment mars 2018 et novembre 2017, où quelques retards positifs apparaissent.
# Cependant, la heatmap montre que ces valeurs viennent presque uniquement de cases avec 1 ou 2 commandes, ce qui les rend peu représentatives.
# À l’inverse, dans les zones où le volume de commandes est élevé, les retards restent très stables et majoritairement négatifs, quel que soit le nombre de versements.
# Cela confirme que les pics extrêmes observés dans le premier graphique ne reflètent pas un comportement réel, mais simplement un manque de données pour les versements élevés.
# Aucun schéma cohérent ne relie la période et le nombre de versements à une hausse des retards.
# Les données indiquent donc que les paiements en plusieurs fois ne provoquent pas de délais particuliers.
# La logistique reste globalement constante dans le temps, et les exceptions visibles proviennent surtout de volumes trop faibles pour en tirer une conclusion fiable.
```

```python
# Conclusion
# L’hypothèse de départ supposait que le mode de paiement ou le nombre de versements pouvait influencer la satisfaction client.
# Les graphiques montrent finalement que ce n’est pas vraiment le cas : les scores restent très proches d’un mode de paiement à l’autre.
# Les clients qui utilisent la carte de crédit paient plus souvent en plusieurs fois, mais cela n’a pas d’impact clair sur leurs avis.
# La seule vraie différence observée concerne le montant du panier : plus il est élevé, plus les avis négatifs deviennent fréquents.
# Cela suggère que la sensibilité des clients vient surtout du prix, et non de la façon dont ils paient.
# On constate aussi que la livraison reste globalement en avance et ne change pas selon le nombre de versements.
# Les pics extrêmes visibles dans certains graphiques viennent surtout de très petits volumes, donc ils ne reflètent pas une tendance réelle.
# Au final, le comportement de paiement ne semble pas être une cause importante d’insatisfaction.
# Les variations d’avis s’expliquent plutôt par d’autres facteurs vus dans les analyses précédentes, comme la période ou la logistique.
# L’hypothèse est donc infirmée, mais l’analyse permet d’identifier ce qui influence réellement la satisfaction.
```
